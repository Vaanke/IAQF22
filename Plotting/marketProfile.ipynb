```python
import pandas as pd
import matplotlib.pyplot as plt

# def sigProfile(bench, label, path):
#     '''
#     bench: pd.series benchmark return data, datetime index
#     laebl: pd.series, datetime index
#     '''
#     bench.fillna(0, inplace=True)
#     label.fillna(0, inplace=True)

#     bench.index = pd.to_datetime(bench.index)
#     label.index = pd.to_datetime(label.index)
#     bname = bench.name
#     lname = label.name
#     labelret = pd.concat([bench,label], axis=1, join = 'inner')
    
#     bench.index = pd.to_datetime(bench.index)
#     label.index = pd.to_datetime(label.index)
#     bname = bench.name
#     lname = label.name
#     labels = label.unique()
#     labelret = pd.concat([bench,label], axis=1, join = 'inner')
    
#     fig, ax = plt.subplots(len(labels), 2, figsize=(10*len(labels), 10))
#     for i, l in enumerate(labels):
#         labelret[labelret[lname]==i][bname].hist(bins=60, ax=ax[i, 0], color = 'lightseagreen', label = 'ss')
#         labelret[labelret[lname]==i][lname].hist(bins=60, ax=ax[i, 1], color = 'b')
#         mean = labelret[labelret[lname]==i][bname].mean()
#         print("mean = ", mean)
#         ax[i, 0].axvline(mean, color="darkorange")
#         ax[i, 0].set_title(bname)
#         ax[i, 1].set_title(lname)
  
#     plt.savefig(path)

#     plt.show()
    
#     return labelret
def sigProfile(bench, label, path):
    '''
    bench: pd.series benchmark return data, datetime index
    laebl: pd.series, datetime index
    '''
    bench.fillna(0, inplace=True)
    label.fillna(0, inplace=True)
    bench.index = pd.to_datetime(bench.index)
    label.index = pd.to_datetime(label.index)
    bname = bench.name
    lname = label.name
    labelret = pd.concat([bench,label], axis=1, join = 'inner')
    
    bench.index = pd.to_datetime(bench.index)
    label.index = pd.to_datetime(label.index)
    bname = bench.name
    lname = label.name
    labels = label.unique()
    labelret = pd.concat([bench,label], axis=1, join = 'inner')
    
    fig, ax = plt.subplots(len(labels), figsize=(5*len(labels), 10))
    for i, l in enumerate(labels):
        labelret[labelret[lname]==i][bname].hist(bins=60, ax=ax[i])
        mean = labelret[labelret[lname]==i][bname].mean()
        print("mean = ", mean)
        ax[i].axvline(mean, color="red")
        ax[i].set_title(bname)
        xlim = ax[i].get_xlim()
        ylim = ax[i].get_ylim()
        ax[i].text(xlim[0] + (xlim[1]-xlim[0])/4*3, ylim[0] + (ylim[1]-ylim[0])/4*3, 
                   f"label={l}\nmean={round(labelret[labelret[lname]==i][bname].mean(), 4)}",
                  size=17)
    plt.savefig(path)
        
    plt.show()
    
    return labelret

def clusterLabel(bench, label):
    '''
    bench: pd.series benchmark return data
    laebl: pd.series
    '''
    
    labelret = pd.DataFrame(label).merge(bench,how='inner' ,left_index=True, right_index=True)
    table = labelret.groupby(label.name).agg(['mean', 'std'])
    return table, labelret
```
